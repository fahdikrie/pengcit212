```python
original = io.imread('cats.jpg')
hsv = color.rgb2hsv(original)
gray = color.rgb2gray(original)
```

```python
plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(original)
plt.subplot(1, 3, 2)
plt.title("HSV")
plt.imshow(hsv)
plt.subplot(1, 3, 3)
plt.title("Gray")
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(original)
plt.subplot(1, 3, 2)
plt.title("HSV")
plt.imshow(hsv)
plt.subplot(1, 3, 3)
plt.title("Gray")
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
hue = hsv[:,:,0]
value = hsv[:,:,2]
plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1)
plt.title("Hue")
plt.imshow(hue, cmap='hsv')
plt.subplot(1, 2, 2)
plt.title("Value")
plt.imshow(value, cmap='gray')
```

```python
value = hsv[:,:,2]
# print(hsv)
# print(hsv.shape)
# print(value.shape)

value = value/2
hsv2 = hsv
hsv2[:,:,2] = value
img = color.hsv2rgb(hsv2)
plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1)
plt.title("Ori")
plt.imshow(original)
plt.subplot(1, 2, 2)
plt.title("V/2")
plt.imshow(img)
```
